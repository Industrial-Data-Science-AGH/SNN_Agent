```python
from pathlib import Path
import shutil

SRC  = Path('/home/umarly-poeta/projects/SNN_Agent/dataset/training')
DST  = Path('/home/umarly-poeta/projects/SNN_Agent/notebooks/dataset')

glass_dir    = DST / 'glass'
negative_dir = DST / 'negative'
glass_dir.mkdir(parents=True, exist_ok=True)
negative_dir.mkdir(parents=True, exist_ok=True)

moved_glass    = []
moved_negative = []
skipped        = []

for f in sorted(SRC.glob('*.wav')):
    name = f.name
    if name.startswith('Glass_Breaking'):
        shutil.move(str(f), glass_dir / name)
        moved_glass.append(name)
    elif name.startswith('Background'):
        shutil.move(str(f), negative_dir / name)
        moved_negative.append(name)
    else:
        skipped.append(name)

print(f'Glass    → {glass_dir}')
for n in moved_glass:
    print(f'  {n}')

print(f'\nNegative → {negative_dir}')
for n in moved_negative:
    print(f'  {n}')

if skipped:
    print(f'\nPominięte (nieznany prefix):')
    for n in skipped:
        print(f'  {n}')

print(f'\nPodsumowanie: {len(moved_glass)} glass, {len(moved_negative)} negative, {len(skipped)} pominiętych')
```
