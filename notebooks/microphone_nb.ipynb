```python
import numpy as np
import scipy.signal as signal
import soundfile as sf
import matplotlib.pyplot as plt
from pathlib import Path

# ─────────────────────────────────────────────────────────────
# KONFIGURACJA
# ─────────────────────────────────────────────────────────────

AFE_CONFIG = {
    "gain":         40.0,
    "hp_freq":      800,
    "lp_freq":      9000,
    "filter_order": 2,
    "env_tau":      0.008,
}

NEURON_CONFIG = {
    "N1": {
        "hp": 1000, "lp": 3000,
        "tau_mem":  0.200,
        "tau_syn":  0.020,
        "v_thresh": 0.22,
        "v_reset":  0.0,
    },
    "N2": {
        "hp": 3000, "lp": 8000,
        "tau_mem":  0.015,
        "tau_syn":  0.005,
        "v_thresh": 0.38,
        "v_reset":  0.0,
    },
    "N3": {
        "tau_mem":  0.400,
        "tau_syn":  0.050,
        "v_thresh": 0.70,
        "v_reset":  0.0,
        "w_N1":     0.15,
        "w_N2":     0.25,
    },
    "N_inh": {
        "hp": 800, "lp": 9000,
        "tau_mem":  0.600,
        "tau_syn":  0.100,
        "v_thresh": 0.30,
        "v_reset":  0.0,
        "w_inh":    0.40,
    },
}

# ─────────────────────────────────────────────────────────────
# BLOK 1: WCZYTANIE AUDIO
# ─────────────────────────────────────────────────────────────

def load_audio(path: str, target_sr: int = 22050):
    audio, sr = sf.read(path, always_2d=False)

    if audio.ndim == 2:
        audio = audio.mean(axis=1)

    if sr != target_sr:
        from scipy.signal import resample_poly
        from math import gcd
        g = gcd(target_sr, sr)
        audio = resample_poly(audio, target_sr // g, sr // g)
        sr = target_sr

    audio = audio.astype(np.float32)
    audio /= (np.max(np.abs(audio)) + 1e-8)

    return audio, sr

# ─────────────────────────────────────────────────────────────
# BLOK 2: AFE
# ─────────────────────────────────────────────────────────────

def apply_gain(audio, gain):
    return np.clip(audio * gain, -1.0, 1.0)


def bandpass_filter(audio, sr, hp_freq, lp_freq, order=2):
    nyq = sr / 2.0
    if hp_freq is not None and lp_freq is not None:
        sos = signal.butter(order, [hp_freq / nyq, lp_freq / nyq],
                            btype='band', output='sos')
    elif hp_freq is not None:
        sos = signal.butter(order, hp_freq / nyq, btype='high', output='sos')
    elif lp_freq is not None:
        sos = signal.butter(order, lp_freq / nyq, btype='low', output='sos')
    else:
        return audio
    return signal.sosfilt(sos, audio)


def envelope_detector(audio, sr, tau):
    rectified = np.maximum(audio, 0.0)
    alpha = np.exp(-1.0 / (tau * sr))
    envelope = np.zeros_like(rectified)
    y = 0.0
    for i, x in enumerate(rectified):
        y = alpha * y + (1.0 - alpha) * x
        envelope[i] = y
    return envelope

# ─────────────────────────────────────────────────────────────
# BLOK 3: NEURON LIF + DETEKTOR
# ─────────────────────────────────────────────────────────────

def simulate_lif(input_signal, sr, tau_mem, tau_syn, v_thresh, v_reset=0.0):
    dt = 1.0 / sr
    n = len(input_signal)
    vmem   = np.zeros(n)
    spikes = np.zeros(n, dtype=bool)
    v = v_reset
    for i in range(1, n):
        I_syn = input_signal[i]
        leak  = v * (dt / tau_mem)
        drive = I_syn * (dt / tau_syn)
        v = max(v - leak + drive, 0.0)
        if v >= v_thresh:
            spikes[i] = True
            v = v_reset
        vmem[i] = v
    return spikes, vmem


def count_triggered(spikes_N3, min_count=2, max_count=24):
    """
    Glass breaking:  N3 = 2–10 spikeów  → trigger
    Background:      N3 = 11–26 spikeów → brak triggera (za dużo)
    Cisza/słabe:     N3 = 0–1           → brak triggera (za mało)
    """
    n = int(spikes_N3.sum())
    return min_count <= n <= max_count

# ─────────────────────────────────────────────────────────────
# BLOK 4: SIEĆ NEURONÓW
# ─────────────────────────────────────────────────────────────

def simulate_network(audio, sr):
    dt = 1.0 / sr
    psp_tau   = 0.010
    alpha_psp = np.exp(-dt / psp_tau)

    sig_N1 = bandpass_filter(audio, sr, NEURON_CONFIG["N1"]["hp"], NEURON_CONFIG["N1"]["lp"])
    sig_N1 = apply_gain(sig_N1, AFE_CONFIG["gain"])
    env_N1 = envelope_detector(sig_N1, sr, AFE_CONFIG["env_tau"])
    spikes_N1, vmem_N1 = simulate_lif(env_N1, sr,
        tau_mem  = NEURON_CONFIG["N1"]["tau_mem"],
        tau_syn  = NEURON_CONFIG["N1"]["tau_syn"],
        v_thresh = NEURON_CONFIG["N1"]["v_thresh"])

    sig_N2 = bandpass_filter(audio, sr, NEURON_CONFIG["N2"]["hp"], NEURON_CONFIG["N2"]["lp"])
    sig_N2 = apply_gain(sig_N2, AFE_CONFIG["gain"])
    env_N2 = envelope_detector(sig_N2, sr, AFE_CONFIG["env_tau"])
    spikes_N2, vmem_N2 = simulate_lif(env_N2, sr,
        tau_mem  = NEURON_CONFIG["N2"]["tau_mem"],
        tau_syn  = NEURON_CONFIG["N2"]["tau_syn"],
        v_thresh = NEURON_CONFIG["N2"]["v_thresh"])

    sig_inh = bandpass_filter(audio, sr, NEURON_CONFIG["N_inh"]["hp"], NEURON_CONFIG["N_inh"]["lp"])
    sig_inh = apply_gain(sig_inh, AFE_CONFIG["gain"])
    env_inh = envelope_detector(sig_inh, sr, 0.05)
    spikes_inh, _ = simulate_lif(env_inh, sr,
        tau_mem  = NEURON_CONFIG["N_inh"]["tau_mem"],
        tau_syn  = NEURON_CONFIG["N_inh"]["tau_syn"],
        v_thresh = NEURON_CONFIG["N_inh"]["v_thresh"])

    w_N1  = NEURON_CONFIG["N3"]["w_N1"]
    w_N2  = NEURON_CONFIG["N3"]["w_N2"]
    w_inh = NEURON_CONFIG["N_inh"]["w_inh"]

    input_N3 = np.zeros(len(audio))
    psp_val  = 0.0
    psp_inh  = 0.0

    for i in range(len(audio)):
        psp_val *= alpha_psp
        psp_val += w_N1 * float(spikes_N1[i]) + w_N2 * float(spikes_N2[i])
        psp_inh *= alpha_psp
        psp_inh += w_inh * float(spikes_inh[i])
        input_N3[i] = max(0.0, psp_val - psp_inh)

    spikes_N3, vmem_N3 = simulate_lif(input_N3, sr,
        tau_mem  = NEURON_CONFIG["N3"]["tau_mem"],
        tau_syn  = NEURON_CONFIG["N3"]["tau_syn"],
        v_thresh = NEURON_CONFIG["N3"]["v_thresh"])

    triggered = count_triggered(spikes_N3)

    return {
        "N1": {"spikes": spikes_N1, "vmem": vmem_N1, "count": int(spikes_N1.sum())},
        "N2": {"spikes": spikes_N2, "vmem": vmem_N2, "count": int(spikes_N2.sum())},
        "N3": {"spikes": spikes_N3, "vmem": vmem_N3, "count": int(spikes_N3.sum())},
        "TRIGGERED": triggered,
    }

# ─────────────────────────────────────────────────────────────
# BLOK 5: EWALUACJA
# ─────────────────────────────────────────────────────────────

def evaluate_dataset(dataset_path: str, max_per_class: int = 50, seed: int = 42):
    """
    max_per_class — ile losowych próbek wziąć z każdej klasy (None = wszystkie).
    seed          — ziarno losowości dla powtarzalności wyników.
    """
    rng = np.random.default_rng(seed)
    results = []
    base = Path(dataset_path)

    for label, true_class in [("glass", True), ("negative", False)]:
        folder = base / label
        if not folder.exists():
            print(f"brak folderu: {folder}")
            continue

        wav_files = sorted(folder.glob("*.wav")) + sorted(folder.glob("*.WAV"))

        if max_per_class is not None and len(wav_files) > max_per_class:
            indices   = rng.choice(len(wav_files), size=max_per_class, replace=False)
            wav_files = [wav_files[i] for i in sorted(indices)]

        print(f"\n[{label.upper()}] {len(wav_files)} próbek (limit={max_per_class})...")

        for wav_path in wav_files:
            audio, sr = load_audio(str(wav_path))
            net = simulate_network(audio, sr)
            predicted = net["TRIGGERED"]

            results.append({
                "file":      wav_path.name,
                "label":     label,
                "true":      true_class,
                "predicted": predicted,
                "n1_spikes": net["N1"]["count"],
                "n2_spikes": net["N2"]["count"],
                "n3_spikes": net["N3"]["count"],
            })

            marker = "✓" if predicted == true_class else "✗"
            print(f"  {marker} {wav_path.name:30s} "
                  f"N1={net['N1']['count']:3d} N2={net['N2']['count']:3d} "
                  f"N3={net['N3']['count']:2d} -> "
                  f"{'TRIGGER' if predicted else 'no trigger'}")

    tp = sum(1 for r in results if r["true"] and r["predicted"])
    fp = sum(1 for r in results if not r["true"] and r["predicted"])
    tn = sum(1 for r in results if not r["true"] and not r["predicted"])
    fn = sum(1 for r in results if r["true"] and not r["predicted"])
    total = len(results)

    if total == 0:
        print("\nBrak danych.")
        return results

    print(f"""
+--------------------------------------+
|          WYNIKI EWALUACJI            |
+--------------------------------------+
|  TP (glass -> TRIGGER):   {tp:4d}       |
|  FP (negative -> TRIGGER):{fp:4d}  !   |
|  TN (negative -> no):     {tn:4d}       |
|  FN (glass -> no trigger):{fn:4d}  !   |
+--------------------------------------+
|  Accuracy:   {(tp+tn)/total*100:5.1f}%               |
|  Precision:  {tp/(tp+fp+1e-8)*100:5.1f}%               |
|  Recall:     {tp/(tp+fn+1e-8)*100:5.1f}%               |
+--------------------------------------+
    """)
    return results

# ─────────────────────────────────────────────────────────────
# BLOK 6: WIZUALIZACJA jednej próbki
# ─────────────────────────────────────────────────────────────

def plot_sample(wav_path: str, title: str = ""):
    audio, sr = load_audio(wav_path)
    net = simulate_network(audio, sr)
    t = np.arange(len(audio)) / sr

    fig, axes = plt.subplots(5, 1, figsize=(14, 10), sharex=True)
    fig.suptitle(f"AFE + SNN pipeline: {title or wav_path}", fontsize=12)

    axes[0].plot(t, audio, color='#4488ff', lw=0.5, alpha=0.8)
    axes[0].set_ylabel("Audio\n(raw)")
    axes[0].set_ylim(-1.1, 1.1)
    axes[0].axhline(0, color='gray', lw=0.3)

    axes[1].plot(t, net["N1"]["vmem"], color='#4488ff', lw=1)
    axes[1].axhline(NEURON_CONFIG["N1"]["v_thresh"], color='red', lw=0.8, ls='--')
    spike_t = t[net["N1"]["spikes"]]
    axes[1].scatter(spike_t, np.ones_like(spike_t) * NEURON_CONFIG["N1"]["v_thresh"] * 1.1,
                    color='#ffaa00', s=20, zorder=5)
    axes[1].set_ylabel(f"N1 V_mem\n({net['N1']['count']} spikow)")

    axes[2].plot(t, net["N2"]["vmem"], color='#00cc88', lw=1)
    axes[2].axhline(NEURON_CONFIG["N2"]["v_thresh"], color='red', lw=0.8, ls='--')
    spike_t = t[net["N2"]["spikes"]]
    axes[2].scatter(spike_t, np.ones_like(spike_t) * NEURON_CONFIG["N2"]["v_thresh"] * 1.1,
                    color='#ffaa00', s=20, zorder=5)
    axes[2].set_ylabel(f"N2 V_mem\n({net['N2']['count']} spikow)")

    axes[3].plot(t, net["N3"]["vmem"], color='#ff6633', lw=1.5)
    axes[3].axhline(NEURON_CONFIG["N3"]["v_thresh"], color='red', lw=0.8, ls='--')
    spike_t = t[net["N3"]["spikes"]]
    axes[3].scatter(spike_t, np.ones_like(spike_t) * NEURON_CONFIG["N3"]["v_thresh"] * 1.1,
                    color='#ffaa00', s=30, zorder=5, marker='*')
    axes[3].set_ylabel(f"N3 V_mem\n({net['N3']['count']} spikow)")

    for row, (neuron, color) in enumerate([("N1", '#4488ff'), ("N2", '#00cc88'), ("N3", '#ff6633')]):
        spk_t = t[net[neuron]["spikes"]]
        axes[4].scatter(spk_t, np.full_like(spk_t, row), color=color, s=8, marker='|', lw=1.5)
    axes[4].set_yticks([0, 1, 2])
    axes[4].set_yticklabels(["N1", "N2", "N3"])
    axes[4].set_ylabel("Raster")
    axes[4].set_xlabel("Czas [s]")

    result = "TRIGGER (RPi WAKE)" if net["TRIGGERED"] else "brak triggera"
    color  = "red" if net["TRIGGERED"] else "green"
    fig.text(0.98, 0.02, result, ha='right', color=color, fontsize=11, fontweight='bold')

    for ax in axes:
        ax.grid(True, alpha=0.2, lw=0.4)
    plt.tight_layout()
    plt.show()


# ─────────────────────────────────────────────────────────────
# URUCHOMIENIE
# ─────────────────────────────────────────────────────────────

evaluate_dataset("dataset/", max_per_class=50)
```
